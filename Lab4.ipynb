```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/dur an/Desktop/Python/Lab/4.Hafta/data.csv")

data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
data.tail()
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]

plt.scatter(M.radius_mean,M.texture_mean,color="red",label="kÃ¶tÃ¼",linewidth=0.3)
plt.scatter(B.radius_mean,B.texture_mean,color="yellow",label="iyi",linewidth=0.3)
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
y=data.diagnosis.values
x_data = data.drop(["diagnosis"],axis=1)
```

```python
import warnings 
warnings.simplefilter(action='ignore',category=FutureWarning)

x = (x_data - np.min(x_data))/ (np.max(x_data) - np.min(x_data))
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print(" {} nn score: {} ".format(9,knn.score(x_test,y_test)))
```

```python
score_list = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

```python
y_pred = knn2.predict(x_test)
y_true = y_test
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true,y_pred)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

f, ax = plt.subplots(figsize=(3,3))
sns.heatmap(cm,annot = True, linewidth=0.3,linecolor="orange",fmt="f",ax=ax )
plt.xlabel('y_true')
plt.ylabel('y_pred')
plt.show()
```
